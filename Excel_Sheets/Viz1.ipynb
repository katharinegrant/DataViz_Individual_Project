```python
import seaborn as sns
import pandas as pd
import numpy as np
%pylab inline
```

```python
state_death = pd.read_csv('State_death.csv')
state_guns = pd.read_csv('State_guns.csv')
state_pop = pd.read_csv('State_pop.csv')
```

```python
state_death.head()
```

```python
state_death.set_index('Location', inplace=True)
state_death.head()
```

```python
state_death = state_death.drop(['United States'], axis=0)
state_death.reset_index(level=0,inplace=True)
state_death.columns = ['State','GunDeathPer100k2016']
state_death.head()
```

```python
state_guns.columns = ['State','Guns2017']
state_guns.head()
```

```python
state_pop.head()
```

```python
state_pop.columns = ['Rank 2017','Rank 2010','State','PopEst2017','Census population 2010','Total HoR seats 2013–2023','Estimated population per House seat, 2017','Census population per House seat, 2010','Percent of the total U.S. population, 2017']
state_pop.head()
```

```python
state_data1 = state_death.merge(state_guns, left_on = 'State', right_on='State')
state_data1.head()
```

```python
state_data1.describe()
```

```python
state_data = state_data1.merge(state_pop, left_on = 'State', right_on='State')
```

```python
state_data.describe()
```

```python
state_data.head()
```

```python
state_data['Guns2017'] = state_data['Guns2017'].apply(pd.to_numeric)
state_data['PopEst2017'] = state_data['PopEst2017'].apply(pd.to_numeric)
state_data['GunDeathPer100k2016'] = state_data['GunDeathPer100k2016'].apply(pd.to_numeric)
state_data['Gun Rate per 100'] = 100*state_data.Guns2017/state_data.PopEst2017
state_data['Type'] = "State"
state_data.describe()
```

```python
state_data.head()
```

```python
state_data.drop(['Guns2017','PopEst2017','Rank 2017','Rank 2010','Census population 2010','Total HoR seats 2013–2023','Estimated population per House seat, 2017','Census population per House seat, 2010','Percent of the total U.S. population, 2017'], axis=1,inplace=True)
state_data.columns = ['Place','GunDeathPer100k','Gun Rate per 100','Type']
state_data.head()
```

```python
country = pd.read_csv('Country_guns_and_death.csv')
HDI = pd.read_csv('Human development index.csv')
```

```python
country.head()
```

```python
HDI.head()
```

```python
country_data = country.merge(HDI, left_on = 'Country', right_on='Country')
country_data.describe()
```

```python
country_data.head()
```

```python
country_data.drop(['Year','Homicides','Suicides','Unintentional','Undetermined','Sources and notes','HDI Rank (2015)'], axis=1,inplace=True)
country_data['Type'] = "Country"
country_data.columns = ['Place','GunDeathPer100k','Gun Rate per 100','HDI','Type']
country_data.head()
```

```python
all_data = pd.concat([country_data, state_data])
```

```python
all_data.head()
```

```python
all_data.to_csv("gun_and_death.csv")
```
